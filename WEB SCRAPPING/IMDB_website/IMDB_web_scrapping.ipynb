```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import numpy as np
```

```python
import time
```

```python
headers = {'Accept-Language':"en-US,en;q=0.5"}
```

```python
pages = np.arange(1,200,100)
```

```python
# movie_name_list = []
# year_list=[]
# time_list=[]
# rating_list =[]
# metascore_list=[]
# votes_list=[]
# gross_list = []
list_movies = []
for page in pages :
    url = requests.get("https://www.imdb.com/search/title/?groups=top_1000&sort=user_rating,desc&count=100&start="+str(page)+"&ref=adv_nxt")
    soup = BeautifulSoup(url.text,'html.parser')
    movie_data = soup.find_all('div',{'class':'lister-item mode-advanced'})
    time.sleep(5)
    for data in movie_data:
        name = data.find('h3',{'class' : 'lister-item-header'}).text
        #movie_name_list.append(name)
        
        year = data.find('span',{'class':'lister-item-year text-muted unbold'}).text
        #year_list.append(year)
        
        runtime = data.find('span',{'class':'runtime'}).text
        #time_list.append(runtime)
        
        rating = data.find('div', {'class':'inline-block ratings-imdb-rating'}).text
        #rating_list.append(rating)
        
#         metascore = data.find('span', {'class':'metascore'}).text
#         metascore_list.append(metascore)
        
        value = data.find_all('span',{'name':'nv'})
        
        vote = value[0].text
        #votes_list.append(vote)
        
        gross = value[1].text if len(value)>1 else '####'
        #gross_list.append(gross)
        dict_movie = {'Movie Name' : name , 'Year of Realease' : year , 'Watch time' : runtime , 'Movie Rating' : rating ,'Votes':vote, 'Gross' : gross}
        list_movies.append(dict_movie)
        
```

```python
import json
with open("imdb2.json","w") as f :     #json ek package(class)   dumps is function
    json.dump(list_movies,f)
```
